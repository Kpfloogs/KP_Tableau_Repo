```python
import pandas as pd
import datetime as dt
from functools import reduce
```

```python
#import 2019 CSVs
may19 = pd.read_csv('Resources/May_Bike_Data_19.csv')
june19 = pd.read_csv('Resources/June_Bike_Data_19.csv')
july19 = pd.read_csv('Resources/July_Bike_Data_19.csv')
aug19 = pd.read_csv('Resources/Aug_Bike_Data_19.csv')
sept19 = pd.read_csv('Resources/Sept_Bike_Data_19.csv')
oct19 = pd.read_csv('Resources/Oct_Bike_Data_19.csv')
```

```python
may19.head()
```

```python
data_2019 = [may19, june19, july19, aug19, sept19, oct19]

full_2019_data = reduce(lambda left, right: pd.merge(left, right, how = "outer"), data_2019)
```

```python
full_2019_data.isnull().sum()
```

```python
data_2019_full = full_2019_data.dropna()
data_2019_full.info()
```

```python
# data_2019_full['starttime'] = pd.to_datetime(data_2019_full['starttime'], format= '%Y-%m-%d %H:%M:%S')
data_2019_full['gender'].replace({0: 'unknown', 1: "male", 2: "female"}, inplace=True)
data_2019_full.head()
```

```python
clean_2019_data = data_2019_full
clean_2019_data.count()
```

```python
clean_unique = clean_2019_data["starttime"].nunique()
clean_unique
```

```python
clean_2019_data.to_csv('output/2019_citi_bike_data.csv')
```
